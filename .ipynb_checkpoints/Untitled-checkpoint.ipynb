```python
citys = [ferrara,torino,mantova,milano,ravenna,asti,bologna,piacenza,cesena,faenza]
for city in citys:
    city.drop('Unnamed: 0',axis=1)
```

显示最高温度于离海远近的关系（观察多个城市）

```python
max_temps=[]
city_dists=[]
for city in citys:
    max_temps.append(city['temp'].max())
    city_dists.append(city['dist'][0])
```

```python
city_dists
```

```python
#plt.figure(figsize=[8,8])
plt.scatter(city_dists,max_temps)
plt.xlabel('海洋距离',fontsize=20)
plt.ylabel('最高温度',fontsize=20)
plt.title('距海距离和温度之间的微妙关系',fontsize=20)
```

观察发现，离海近的可以形成一条直线，离海远的也能形成一条直线。

分别以100公里和50公里为分界点，划分为离海近和离海远的两组数据(近海：小于100  远海：大于50)

```python
np_city_dists=np.array(city_dists)#所有城市距海距离
np_max_temps=np.array(max_temps) #所有城市最大温度
```

```python
far_condition=np_city_dists > 50 #远海城市条件
near_condition=np_city_dists < 100 #近海城市条件
```

```python
#获得近海城市距离和最高温度的数据值
near_max_temps=np_max_temps[near_condition]
near_city_dists=np_city_dists[near_condition]

#获得远海城市距离的最高温度和距离
far_max_temps=np_max_temps[far_condition]
far_city_dists=np_city_dists[far_condition]
```

```python
plt.scatter(near_city_dists,near_max_temps,s=100)
plt.scatter(far_city_dists,far_max_temps)
```

```python
#使用线性回归模型来预测线性规律:就是使用LinearRegression来帮我们画出散点图中的线
from sklearn.linear_model import LinearRegression
```

```python
#构建线性回归模型对象
#假设存在一个函数关系：f(x)=ax+b
linear=LinearRegression()
```

```python
#city_dists是x轴，表示离海距离。max_temps是y轴，表示城市最高温度
#将x，y带入fit中的过程，称之为训练（本质上就是解方程，求解的是一个估算值）
#city_dists和max_temps就是数据样本，多个数据样本组成样本集
#city_dists是特征，max_temps是结果。特征必须多列组合的方式展现，因为特征会有多种
linear.fit(near_city_dists.reshape(-1,1),near_max_temps)
```

```python
#求值
linear.predict(300)
```

```python
x_n=np.linspace(near_city_dists.min(),near_city_dists.max(),100).reshape(-1,1)
y_n=linear.predict(x_n)
#x_f=np.linspace(far_city_dists.min(),far_city_dists.max(),100).reshape(-1,1)
#y_f=linear.predict(x_f)
```

```python
plt.plot(x_n,y_n)
#plt.plot(x_f,y_f)
plt.scatter(near_city_dists,near_max_temps,s=100)
plt.scatter(far_city_dists,far_max_temps)
```
